```python
import numpy as np
import matplotlib.pyplot as plt

import topfarm  # test importu

x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

plt.figure(figsize=(6, 4))
plt.scatter(x, y)
plt.title("Layout turbin – Zajęcia 1")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True)
plt.axis("equal")
plt.show()
```

```python
# Wykres z kolorami i etykietami
colors = ['blue', 'red', 'green', 'yellow']
labels = ['T1', 'T2', 'T3', 'T4']

plt.figure(figsize=(6, 4))
for xi, yi, c, lab in zip(x, y, colors, labels):
    plt.scatter(xi, yi, color=c, s=80)
    plt.text(xi + 20, yi + 20, lab, fontsize=9)

plt.title("Layout turbin – Zajęcia 1 z kolorami i etykietami")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True)
plt.axis('equal')
plt.show()
```
